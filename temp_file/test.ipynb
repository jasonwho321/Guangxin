```python
import pandas as pd
import pyodbc
from sqlalchemy import create_engine
import json
import requests
def read_config(config_file_path):
    with open(config_file_path, 'r') as f:
        config_data = json.load(f)
    return config_data
# 读取配置文件
config_file_path = '/Users/huzhang/PycharmProjects/Guangxin/config.json'  # 根据实际情况调整路径
config_data = read_config(config_file_path)

def download_transportInfo():
    url = "https://chinawms.houseinbox.com/rbcs/transport/exportData"
    headers = {'userid': '958'}
    payload = {
        "transportQuery": {
            "ainstitutionAids": [736, 961, 929, 962, 450, 998, 298, 939, 332, 976, 977, 375, 633, 765, 605]
        }
    }

    # 发送请求
    response = requests.post(url, json=payload, headers=headers)
    print(response.content)
    # 检查响应状态
    if response.status_code == 200:
        # 将数据读入pandas DataFrame
        df = pd.read_excel(response.content,engine='openpyxl')
        df = df[(df['业务类型'] == '电商') & (df['柜号'].notnull()) & (df['柜号'].str.len() == 11)]

        # 将DataFrame保存为.xlsx文件
        df.to_excel("/Users/huzhang/Downloads/transportInfo.xlsx", encoding="gb18030", index=False)
        print("文件已成功保存为 transportInfo.xlsx")
        return df
    else:
        print("请求失败，状态码：", response.status_code)

def upload_to_sql_server(df, table_name, timestamp_column, if_exists='replace'):
    # 可以选择'replace'或'append_no_duplicates'
    # 从配置文件中获取所需的值
    server = config_data['server']
    database = config_data['database']
    user = config_data['user']
    password = config_data['password']
    # 创建数据库连接字符串
    conn_str = (
        f"Driver={{ODBC Driver 17 for SQL Server}};"
        f"Server={server};"
        f"Database={database};"
        f"UID={user};"
        f"PWD={password};"
    )

    # 创建SQLAlchemy引擎
    engine = create_engine(f"mssql+pyodbc:///?odbc_connect={conn_str}")

    # 连接到数据库
    conn = pyodbc.connect(conn_str)

    if if_exists == 'replace':
        df.to_sql(table_name, engine, if_exists="replace", index=False)
        print(f"Data uploaded to table '{table_name}' in database '{database}' on server '{server}'.")
    elif if_exists == 'append_no_duplicates':
        # 获取上传表格中的时间范围
        min_timestamp = df[timestamp_column].min()
        max_timestamp = df[timestamp_column].max()

        # 删除与上传表格中的时间范围重叠的数据
        query = f"DELETE FROM {table_name} WHERE {timestamp_column} >= '{min_timestamp}' AND {timestamp_column} <= '{max_timestamp}'"
        conn.execute(query)
        conn.commit()

        # 将新数据追加到数据库表格中
        df.to_sql(table_name, engine, if_exists="append", index=False)
        print(f"Data uploaded to table '{table_name}' in database '{database}' on server '{server}' without duplicates.")
    else:
        print(f"Error: Invalid 'if_exists' value. Allowed values are 'replace' and 'append_no_duplicates'.")

    # 关闭数据库连接
    conn.close()


table_name = 'Transport.transportInfo_WMS'

timestamp_column = 'ETD'

df = download_transportInfo()
if len(df):
    upload_to_sql_server(df, table_name, timestamp_column)
```
