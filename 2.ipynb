```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://gd.huatu.com/gxlx/kaoshi/2.html'
session = requests.session()
req_header = {
    'User-Agent':'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/67.0.3396.99 Safari/537.36',
}

response = session.get(url,headers=req_header)
response.encoding = 'gbk'
print(response.text)
```

```python
soup = BeautifulSoup(response.text)
```

```python
h3_list = soup.find_all(name='h3')
title_link_dic = {}
for h in h3_list:
    title_link = h.a.attrs
    link = title_link['href']
    title = title_link['title']
    title_link_dic[title] = link
print(title_link_dic)
```

```python
import jieba
```

```python
import jieba.posseg as pseg
total_title_list = []
for t in title_link_dic:
    words = pseg.cut(t)
    title_list = [t]
    for word ,flag in words:
        if flag == 'm' and word != '2021'and word != '2020'and word != ''and word != '':
            title_list.append(word)
        if flag == 'ns':
            title_list.append(word)
    total_title_list.append(title_list)
print(total_title_list)
```
